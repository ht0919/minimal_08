## ■[問題6] データの分割
- 変数irisの行数と列数を表示してください。
- 変数irisの行数の半分の値を計算して変数nに代入し、表示してください。
- 変数iris_trainに変数irisの前半半分を分割して代入してください。
- 変数iris_testに変数irisの後半半分を分割して代入してください。
- 変数iris_trainと変数iris_testの行数と列数をそれぞれ表示してください。

```python
# 分割前のデータの確認
print('元データの行数と列数', iris.shape)
n = iris.shape[0] // 2 # 半分の値を計算(小数切り捨て)
print('半分の値', n)

# データを2つに分割
iris_train = iris.iloc[:n,:] # 前半半分を学習用データとして確保
iris_test  = iris.iloc[n:,:] # 後半半分を検証用データとして確保

# 分割後のデータの確認
print('学習用データの行数と列数', iris_train.shape)
print('検証用データの行数と列数', iris_test.shape)
```

## ■[問題7] データの集計
- 野菜の売上データから種類別の合計と平均をそれぞれ表示してください。

```python
# 野菜の売上データ
df = pd.DataFrame({
    'vegetable': ['cabbage','radish','cabbage','onion','radish','cabbage'],
    'price':     [     128 ,    285 ,     320 ,   168 ,    228 ,     265 ]})

print('野菜の種類別の合計', df.groupby(['vegetable']).sum())
print('野菜の種類別の平均', df.groupby(['vegetable']).mean())
```

## ■[問題8] データの加工1(欠損値の置換)
- 変数titanicに、seabornのデータセットからタイタニックのデータを読み込んでください。
- 変数ageに、変数titanicの年齢(age)の列データをコピーしてください。
- 変数ageの加工前の欠損値の個数を表示してください。
- 変数ageの年齢の最頻値を表示してください。
- 変数ageの欠損値を最頻値で置換してください。
- 変数ageの加工後の欠損値の個数を表示してください。

```python
titanic = sns.load_dataset('titanic')
age = titanic['age'].copy() # 年齢のデータをコピー
print('欠損値の個数：', age.isnull().sum())
print('年齢の最頻値：', age.mode().iloc[0]) # 欠損値を除く最頻値の算出
age = df.fillna(age.mode().iloc[0])         # fillnaで欠損値(NaN)を最頻値で置換
print('欠損値の個数：', age.isnull().sum())
```

## ■[問題9] データの加工2(非数値をダミー変数に変換)
- 変数clsに、変数titanicから階級(class)のデータをコピーしてください。
- 変数clsの変換前のデータを5行表示してください。
- 変数clsのダミー変数を5行表示してください。
- 変数clsをダミー変数に変換してください。
- 変数clsの変換後のデータを5行表示してください。

```python
cls = titanic['class'].copy()     # 階級(cls)のデータをコピー
print(cls.head())                 # 変換前のデータを5行表示
print(pd.get_dummies(cls).head()) # ダミー変数を5行表示
cls = pd.get_dummies(cls)         # ダミー変数に変換
print(cls.head())                 # 変換後のデータを5行表示
```
